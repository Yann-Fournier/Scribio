```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Ouvrir une image
img = Image.open("./img/A.png") # 300 x 300

# Redimensionner l'image
# nouvelle_taille = (300, 300) # image actuelle mais c mieux pour les images avec des détails
nouvelle_taille = (28, 28)
img_redimensionnee = img.resize(nouvelle_taille)

# Convertir l'image en noir et blanc
NB_image = img_redimensionnee.convert('L')

# Sauvegarder l'image redimensionnée
NB_image.save("./img/A_redim2.png")
```

```python
# Ouvrir une image
img = Image.open("./img/Chien_redim2.png") # 300 x 300

tab = []

for i in range(img.size[1]):
    row = []
    for y in range(img.size[0]):
        # Récupérer la couleur du pixel
        row.append(img.getpixel((y, i)))
    tab.append(np.array(row))
    
tab_numpy = np.array(tab)
tab_numpy
```

```python
plt.matshow(tab_numpy) # faire apparaitre l'image grace à matplotlib
```
